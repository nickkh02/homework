### Задание №1

Скачайте текст "Литературных анекдотов". Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.

```python
# Импортирую необходимые библиотеки
from pymystem3 import Mystem
import re 

# Называю функцию, которая должна будет брать два аргумента
def lemm_master(input_file, output_file):
    # эта часть функции отвечает за очистку input_file от лишних знаков
    f = open(input_file, encoding='utf-8')
    anec = f.read()
    anec_punct_alpha = (re.sub('[,.":?!)(;]', ' ', anec))
    anec_punct = (re.sub('- | -', '', anec_punct_alpha))
    # эта часть функции отвечает за лемматизацию текста 
    m = Mystem()
    lemmas = m.lemmatize(anec_punct)
    # наконец, эта часть функции создает новый файл и построчно копирует файл
    with open(output_file, 'a+') as p: 
        for f in lemmas:
            p.write(f)

# вызываю функцию, задаю ей значения
lemm_master('literary_anecdotes.txt', 'lemmas.txt')

# проверяю результат на правильность
with open('lemmas.txt', 'r') as g: 
    lmm = g.read()
    print(lmm)
```

3\. Пройдитесь циклом по списку с разборами, который выдал pymystem3, извлекая из каждого разбора форму слова и его часть речи и записывая их в новый словарь (форма -- ключ, часть речи -- значение). 

```python
# Импортирую модуль регулярных выражений
import re

# В этот словарь будут записываться результаты разбора
my_dict = dict()

# Решил сократить средства для поиска части речи и формы
forma = re.compile(r"'text': '(.*?)'")
chast = re.compile('[A-Z]+')

# Для корректной работы re.sub выгружаю содержимое файла в простую переменную
f = open('dict.txt', 'r', encoding='utf-8')
# использую docs quotes, иначе re.sub будет работать немного не так, как надо
skibidi = ''''''
for line in f:
    skibidi = skibidi + line
f.close()
f.closed

# Вычищаю текст от лишней информации, которая может помешать парсингу
rapapa = (re.sub((r".'text': '[^а-яА-Я]+'}.|{'analysis'.*?'wt'.*?(?<!\.)\d+(?!\.), "), "", skibidi))

# Перевожу отредактированный текст в список для пущего удобства
multi = (re.split(r"},", rapapa))

# Целый час ломал голову, почему нельзя проставить названия переменных в цикле (пока не вспомнил про global/local scope)
one_list = []
two_list = []

# Начинаю цикл
for line in multi:
    # Резульатам поиска каждой из частей словаря присваивается по своему названию
    match1 = forma.findall(line)
    match2 = chast.findall(line)
    # Если в теге есть и слово, и часть речи, то результаты сводятся в соответствующие списки
    if match1 and match2:
        one_list = one_list + match1
        two_list = two_list + match2
    # В случае, если такого не происходит, цикл ничего не записывает и идет дальше
    else:
        pass
    # Итоговый результат записывается в словарь
    my_dict = dict(zip(one_list, two_list))

# Визализирую результат
print("Содержимое словаря:", my_dict)
```

4\. Посчитайте абсолютную частоту для всех частей речи.

```python
# Нахожу все части речи в очищенном от лишних знаков тексте
chasts = chast.findall(rapapa)

# Перевожу в строку для для использования функции Counter
chasts_str = str(chasts)

# Подсчитываю в Counter число (предварительно разделив с помощью split)
chasts_str_count = Counter(chasts_str.split())

# Вывожу на экран абсолютую частоту всех частей речи (из-за особенностей регулярного выражения "потерялось" одно причастие и глагол)
print(chasts_str_count)
```

```python
# Делаю пустой словарь, в котором будут храниться данные по относительной частоте
abs_cha = {}

# Удаляю из счетчика лишние элементы
del chasts_str_count[ "['PR',"]
del chasts_str_count[ "'V']"]  

# Для большей точности прибавляю эти значения к их товарищам
chasts_str_count[ "'PR',"] += 1
chasts_str_count[ "'V',"] += 1

# С помощью этого цикла я смогу посчитать относительную частоту 
for c, n in chasts_str_count.items():
    # Числа из счетчика я делаю на общую длину текста
    cha = n/len(chasts_str)
    # Получившийся результат отношу к одному из ключей
    abs_cha[c] = cha
    
# Демонстрирую результат    
print(abs_cha)
```
