### Задание №1

Скачайте текст "Литературных анекдотов". Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.

```python
# Импортирую необходимые библиотеки
from pymystem3 import Mystem
import re 

# Называю функцию, которая должна будет брать два аргумента
def lemm_master(input_file, output_file):
    # эта часть функции отвечает за очистку input_file от лишних знаков
    f = open(input_file, encoding='utf-8')
    anec = f.read()
    anec_punct_alpha = (re.sub('[,.":?!)(;]', ' ', anec))
    anec_punct = (re.sub('- | -', '', anec_punct_alpha))
    # эта часть функции отвечает за лемматизацию текста 
    m = Mystem()
    lemmas = m.lemmatize(anec_punct)
    # наконец, эта часть функции создает новый файл и построчно копирует файл
    with open(output_file, 'a+') as p: 
        for f in lemmas:
            p.write(f)
    f.close()

# вызываю функцию, задаю ей значения
lemm_master('literary_anecdotes.txt', 'lemmas.txt')

# проверяю результат на правильность
with open('lemmas.txt', 'r') as g: 
    lmm = g.read()
    print(lmm)
```

3\. Пройдитесь циклом по списку с разборами, который выдал pymystem3, извлекая из каждого разбора форму слова и его часть речи и записывая их в новый словарь (форма -- ключ, часть речи -- значение). 

```python
# Импортирую модуль регулярных выражений
import re

# В этот словарь будут записываться результаты разбора
my_dict = dict()

# Решил сократить средства для поиска части речи и формы
forma = re.compile(r"{'text': '(.*?)'")
chast = re.compile('[A-Z]+')

# Для корректной работы re.sub выгружаю содержимое файла в простую переменную
f = open('dict.txt', 'r', encoding='utf-8')
# использую docs quotes, для работы моей функции это важно
skibidi = ''''''
for line in f:
    skibidi = skibidi + line
f.close()
f.closed

# Вычищаю текст от лишней информации, которая может помешать парсингу
ra = (re.sub((r"{'text': '([^А-Я])+}"), "", skibidi))
papa = (re.sub((r"('analysis'.*?)'gr'"), "", ra))

# С помощью этого цикла прохожусь по списку
for line in papa:
    # Назначаю переменные для результатов поиска по одной из частей словаря
    match1 = forma.findall(papa)
    match2 = chast.findall(papa)
    # Если результаты поиска успешны, им назначаются еще переменные 
    if match1: 
        one_tuple = match1
    if match2:
        two_tuple = match2
    # Эти переменные сводятся в словарь при помощи этой нехитрой комбинации
    my_dict = dict(zip(one_tuple, two_tuple))
        
# Визализирую результат
print("Содержимое словаря:", my_dict)
```
